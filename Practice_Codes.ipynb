### 1. Reverse a String (Without using built-in)

```python
def reverse_string(s):
    rev = ""
    for char in s:
        rev = char + rev
    return rev

print(reverse_string("python"))
```

### 2. Check Palindrome String

```python
def is_palindrome(s):
    return s == s[::-1]

print(is_palindrome("madam"))
print(is_palindrome("python"))
```

### 3. Count Vowels and Consonants

```python
vowel = ['A','E','I','O','U','a','e','i','o','u']
word = "Python Programming"
count1 = 0
count2= 0
for character in word:
    if character in vowel:
        count1+=1
    if character not in vowel:
        count2+=1
print(count1)
print(count2)
```

### 4. Frequency of Characters in a String

```python
word = "programming"
count = 0
character = "o"
for i in word:
    if i == character:
        count+=1

print(count)
```

### 5. Fibonacci Series Using Recursion

```python
def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n - 1) + fibonacci(n - 2)

for i in range(7):
    print(fibonacci(i), end=" ")
```

### 6. Check Armstrong Number

```python
def is_armstrong(num):
    digits = str(num)
    power = len(digits)
    total = sum(int(d)**power for d in digits)
    return total == num

print(is_armstrong(153))
print(is_armstrong(123))
```

### 7. Find Second Largest Element in List

```python
numbers = [10, 20, 4, 45, 99]

numbers = list(set(numbers))   # remove duplicates
numbers.sort()

print(numbers[-2])
```

### 8. Factorial Using Recursion

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)

print(factorial(5))
```

##  Pattern Printing

```python
# Print Solid Square
''' * * * * 
    * * * *
    * * * *
    * * * *
'''
n = 4  # size of the square

for i in range(n):
    for j in range(n):
        print("*", end=" ")
    print()


```

```python
# Print Hollow Square
''' * * * * 
    *     *
    *     *
    * * * *
'''
n = 4 
for i in range(n):
    for j in range(n):
        if (i==0 or i == n-1 or j==0 or j==n-1):
            print("*",end=" ")
        else:
            print(" ",end=" ")
    print()
```

```python
# Print Right Triangle
''' *
    * *
    * * *
    * * * *
'''
n = 5  # number of rows

for i in range(1, n+1):
    for j in range(i):
        print("*", end=" ")
    print()

```

```python
# Print Inverted Right Triangle
''' * * * *
    * * *
    * * 
    * 
'''

n = 5
for i in range(n):
    for i in range(i,n-1):
        print("*",end=" ")
    print()
```

```python
# Print Inverted Right-Aligned Triangle
''' * * * *
      * * *
        * *
          *
'''

n = 4  # number of rows

for i in range(n, 0, -1):
    print("  " * (n - i), end="")  # spaces
    print("* " * i)                # stars
```

```python
# Print Pyramid
'''    *
      * *
     * * *
    * * * *

'''
n = 5  # number of rows

for i in range(1, n+1):
    # print spaces
    print(" " * (n - i), end="")
    # print stars
    print("* " * i)

```
